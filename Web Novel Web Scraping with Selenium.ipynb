# Web Scraping with Selenium

This is the first part of my project on Japanese web novel titles. In this notebook, I am going to use the selenium package to collect data from Syosetsu.com, the largest web novels platform in Japan.


```python
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
import pandas as pd
```

```python
sort_by = ('weekly','favnovelcnt','hyoka','hyokacnt','yearlypoint','monthlypoint')
url_p1 = 'https://yomou.syosetu.com/search.php?word=&notword=&genre=&type=re&mintime=&maxtime=&minlen=&maxlen=&min_globalpoint=&max_globalpoint=&minlastup=&maxlastup=&minfirstup='
url_p2 = '%2F01%2F01&maxfirstup='
url_p3 = '%2F12%2F31&order='
```

```python
temp_list = []
driver = webdriver.Chrome(executable_path = 'C:/Users/eva/Desktop/python/chromedriver.exe')

for year in range(2007,2022):
    for s in sort_by:
        url = url_p1 + str(year)+ url_p2 + str(year) + url_p3 + s
        driver.get(url)
        for num in range(99):
            retries = 1
            while retries <= 5:
                try:
                    results = driver.find_elements_by_class_name('searchkekka_box')
                    break
                except TimeoutException:
                    driver.refresh()
                    retries += 1
            results = driver.find_elements_by_class_name('searchkekka_box')
            for i in results:
                temp = i.text
                temp = temp + '\n' + str(year)
                temp_list.append(temp.split('\n'))
            retries = 1
            while retries <= 5:
                try:
                    driver.find_element_by_class_name('nextlink').click()
                    break
                except:
                    driver.refresh()
                    retries += 1
            
driver.quit()
```

```python
data = pd.DataFrame(temp_list,columns = ['A','B','C','D','E','F','G','H','I','J','K','L','M','N'])
data.head()
```

```python
df = data.drop_duplicates(subset=['A'], keep='first')
print('The number of observation is',len(df.index))
```

## Data Cleaning

```python
df = df.drop(columns=['C','D','L'])
df = df.rename(columns={'A':'Title','E':'Complete','F':'Parts','G':'Abstract','H':'Genre','N':'Year'})

for i in list(df):
    df[i] =df[i].replace(',', '',regex=True)
    df[i] =df[i].replace('／', '／',regex=True)
```

```python
df[['Author','B','ID','B1']] = df['B'].str.split('／',expand=True)
df['Latest_Update'] = df['J'].str.extract('(\d+)')
df[['F','Word_Count']] = df['J'].str.split('分',expand=True)
df[['Weekly_Viewers','Reviews']] = df['K'].str.split('レビュー数',expand=True)
df[['Bookmarks','Rating','Total_Points']] = df['M'].str.split('人',expand=True)
df['Keywords'] = df['I'].replace('キーワード：','',regex=True)
df['Author'] = df['Author'].replace('作者：','',regex=True)
df['ID'] = df['ID'].replace('Nコード：','',regex=True)
df = df.drop(columns=['B','B1','F','I','J','K','M'])

for i in ('Word_Count','Parts','Weekly_Viewers','Reviews','Bookmarks','Rating','Total_Points'):
    df[i] = df[i].str.extract('(\d+)')

for i in ('Parts','Word_Count','Weekly_Viewers','Reviews','Rating','Total_Points'):
    df[i] = pd.to_numeric(df[i])

for i in ('ジャンル：','〕'):
    df['Genre'] = df['Genre'].replace(i, '',regex=True)
df[['Sub_Genre','Genre']] = df['Genre'].str.split('〔',expand=True)

df['Title_Length'] = df['Title'].str.len()
df['Abstract_Length'] = df['Abstract'].str.len()
df['Average_Length'] = df['Word_Count']/df['Parts']
df['Average_Score'] = df['Total_Points']/df['Rating']
df['Average_Length'] = df['Average_Length'].round(1)
df['Average_Score'] = df['Average_Score'].round(4)

df.head()
```

```python
df = df.replace({
    'Complete': {'連載中': 'On Going','完結済':'Completed','短編':'Oneshot'},
    'Genre': {'恋愛': 'Romance', 'ファンタジー': 'Fantasy','文芸':'Literature',
              'SF':'Sci-fi','その他':'Others','ノンジャンル':'Non-genre'},
    'Sub_Genre': {'異世界': 'Fantasy World','現実世界':'Real World','ハイファンタジー':'High Fantasy',
                  'ローファンタジー':'Low Fantasy','純文学':'Pure Literature','ヒューマンドラマ':'Drama',
                  '歴史':'Historical Fiction','推理':'Mystery','ホラー':'Horror','アクション':'Action Fiction',
                  'コメディー':'Comedy','VRゲーム':'Virtual Reality','宇宙':'Space','空想科学':'Science Fantasy',
                  'パニック':'Disaster Fiction','童話':'Fairy Tale','詩':'Poetry','エッセイ':'Essay','リプレイ':'TRPG',
                  'その他':'Others','ノンジャンル':'Non-genre'}
})
df.head()
```
